### 13- Spliting the strings into parts

```python
import pandas as pd

df = pd.DataFrame({"Name":["Sumit Sharma","Mahadev Khokhar","Parkash Kumar"],
                   "Location":["Tandojam, Pakistan","Hyderabad, Pakistan","Mithi, Pakistan"]})
df
```

```python
# spliting two columns
df[["fname","lname"]] = df.Name.str.split(" ",expand=True)
df[["City","Country"]] = df.Location.str.split(",",expand=True)
# droping columns
df = df.drop(columns=['Name','Location'])
df
```

### 20- Reshaping a dataframe

```python
fasla  = pd.DataFrame([["12345",100,200,300],["67890",400,500,600],["01283",700,800,900]],columns=['zip','factory','warehouse','retail'])
fasla
```

```python
fasla2 = pd.DataFrame([[1,'12345','factory'],[2,'67890','warehouse']],columns=['user_id','zip','location'])
fasla2
```

```python
fasla_long = fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
```

```python
import seaborn as sns

sns.barplot(x='zip',y='distance',hue='location_type',data=fasla_long)
```
