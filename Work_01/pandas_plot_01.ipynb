```python
# all imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
kashti.info()
```

```python
# checking all the null values
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',data=kashti)
```

```python
sns.countplot(x='survived',hue='sex',data=kashti)
```

```python
plt.scatter(data=kashti,x='age',y='fare')
```

```python
sns.scatterplot(data=kashti,x='age',y='fare',hue='pclass')
```

```python
sns.barplot(data=kashti,x='sex',y='fare',hue='pclass')
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist()
```

```python
kashti.hist(bins=10,figsize=(9,7),grid=False)
```

```python
sns.boxplot(data=kashti,x='sex',y='fare',hue='pclass')
```

```python
sns.boxenplot(data=kashti,x='sex',y='fare',hue='pclass' )
```

```python
sns.stripplot(data=kashti, x='sex', y='fare', hue='pclass', palette={1: 'blue', 2: 'seagreen', 3: 'gold'})
plt.title("Gender vs Fare by Passenger Class")
plt.show()
```

```python

# Drop non-numeric columns
numeric_kashti = kashti.select_dtypes(include='number')

# Calculate the correlation matrix
corr = numeric_kashti.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 10))

# Draw the heatmap with annotations
sns.heatmap(corr)

# Add a title to the heatmap
plt.title('Correlation Heatmap', fontsize=16)

# Show the plot
plt.show()
```

```python
sns.violinplot(data=kashti,x='sex',y='fare',hue='pclass')
```
