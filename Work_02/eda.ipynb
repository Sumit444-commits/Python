# Exploratory Data Analysis
Exploratory data analysis (EDA) is a method used to `analysis` and `summarize` datasets. In python, you can use  libraries like `pandas` to perform EDA.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
# finding null values
df.isnull().sum()
```

```python
# finding the percentage of null values
df.isnull().sum() * 100 / len(df)
```

```python
# finding unique values
df.nunique()
```

```python
# columns names
df.columns
```

Viusalizing

```python
# visualizing nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis') # cbar is for color bar on right side
```

```python
df2 = df.copy()
```

```python
df2.drop(columns='deck',inplace=True)
```

```python
df2.columns
```

```python
# visualizing and exploring
sns.countplot(data=df2,x='survived') # In countplot y axis is reserved for count
```

```python
sns.barplot(data=df, x= 'survived', y= 'age', hue='who')
```

```python
sns.countplot(x='pclass',data=df2 , hue='pclass')
```

```python
# data distribution
sns.histplot(data=df2,x='age',bins=10) # bins are the number of bars
# histplot is made when you have numerical data
```

```python
# another way to create histogram
df2['age'].hist(bins=10, color='darkred',alpha=0.75) # alpha is for opacity
```

```python
sns.boxplot(data=df2,y='age',x='survived', hue='who') # when making boxplot one value should be numerical and the other should be categorical
# when making boxplot, x is the column that we want to see the distribution of, y is the column that we want to compare with and hue is the column that we want to see the distribution of x and y in terms of hue
```

```python
# scatter plot is used to see the relationship between two numerical columns
sns.scatterplot(data=df2,x='age',y='fare')
```

```python
sns.scatterplot(data=df2,x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df,hue='survived')
```

```python
df2.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
sns.swarmplot(data=df2,x='survived',y='fare')
```

```python
sns.swarmplot(data=df2,x='survived',y='age',hue='who')
```

```python
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df2)
```

```python
    sns.lmplot(data=df2,x='age',y= (df2['fare']<300).astype(int))
```

```python
df3=df2.copy()
```

```python
# making groups of age
df3['age_group'] = pd.cut(df3['age'],bins=[0,10,18,30,50,80],labels=['Kid','Child','Youth','Middle aged','Old'])
```

```python
df3.head()
```

```python
df3.groupby('age_group')['survived'].mean().plot(kind='bar')
```

```python
sns.pairplot(data=df3,hue='survived',vars=['age','fare','pclass'])
```
